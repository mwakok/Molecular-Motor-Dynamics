## Analyse simulated motor profiles


```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.gridspec as gridspec
import pandas as pd
import pickle
import os
```

```python
# Location of data folder (may contain multiple datasets in multiple subfolders)
data_path = 'G:\\DATA\\SIM_RUNS'     
```

```python
# Find all files within the data folder
files_all = [os.path.join(r,file) for r,d,f in os.walk(data_path) for file in f]

# Unpickle all .data files
RESULTS = pd.DataFrame()
print('Imported files:')
for file in files_all:
    if '.data' in file:                
        infile = open(file,'rb')
        df = pickle.load(infile)
        RESULTS = RESULTS.append(df)
        infile.close
        print(file)   
```

```python
# Option to convolve the data with an experimentally determined PSF
step_size = 24

# Microscope characteristics
fwhm = 700  # width of the experimental Point Spread Function (in nm)
sigma = (fwhm / 2.335) / step_size

# Construct Gaussian filter
sz = 100
x = np.arange(-sz/2, (sz+1)/2)
gauss_filter = np.exp(-x**2 / (2*sigma**2))
gauss_filter = gauss_filter / np.sum(gauss_filter)  # normalize filter
```

### Plot steady-state dynein profiles for a single dynein concentration over multiple lattice sizes

```python
print('Available concentrations for SP dynein: ' + str(np.sort(RESULTS[RESULTS.Type.eq('SP')].Concentration.unique())))
print('Available concentrations for WT dynein: ' + str(np.sort(RESULTS[RESULTS.Type.eq('WT')].Concentration.unique())))
```

```python
# Select single dynein concentration and type to plot
dynein_conc = 1.5
dynein_type = 'SP'
# Option to convolve the dynein profiles with an experimental PSF
convolve_data = True
```

```python
# Setup figure
fig = plt.figure(constrained_layout=True, figsize=(15 ,10))
spec = fig.add_gridspec(ncols=2, nrows=2)

data_filter = RESULTS[RESULTS.Concentration.eq(dynein_conc) & RESULTS.Type.eq(dynein_type)]   
lengths = data_filter.Length.unique()
colors = [cm.copper(x) for x in np.arange(0, 256, len(lengths))]
Y_max = []
Y_width = []

ax = fig.add_subplot(spec[0, :])

# Loop over all lattice sizes
for length, color in zip(lengths, colors):        
    
    # Prepare data
    data_set = RESULTS[RESULTS.Concentration.eq(dynein_conc) & 
                       RESULTS.Length.eq(length) & 
                       RESULTS.Type.eq(dynein_type)]            
    final_state = []
    for data, time in zip(data_set.Lattice, data_set.Frametime):                 
        if time == 0:
            final_state.append(data)
        elif time > 0:
            final_state.append(data[-1])        
    
    if convolve_data:
        Y = np.mean(final_state, axis=0)   
        Y = np.concatenate((np.zeros(sz), Y, np.zeros(sz)), axis=None) 
        Y = np.convolve(Y, gauss_filter, 'same')
        X = np.arange(-sz, len(Y)-sz, 1) * (step_size/1000) 
    else:
        Y = np.mean(final_state, axis=0)  
        X = np.arange(0, len(Y), 1) * (step_size/1000)                
    
    Y_max.append(np.max(Y))
    
    # Calculate width (FWHM) of the traffic jam
    xvals = np.linspace(0, X[-1], len(X)*50)
    Y_interp = np.interp(xvals, X, Y)  
    fwhm_peak = np.where(Y_interp >= Y_max[-1]*0.5)[0]  
    Y_width.append(xvals[fwhm_peak[-1]] - xvals[fwhm_peak[0]])
    
    ax.plot(X, np.flip(Y), color=color, linewidth=2)        

ax.set_xlim((-2, 10))        
ax.set_xlabel('Microtubule length ($\mu$m)', fontsize=14)
ax.set_ylabel('Motor density (a.u.)' , fontsize=14)
ax.text(9, np.max(Y_max), str(dynein_type) + ': ' +str(dynein_conc) + ' nM', 
        horizontalalignment='right', verticalalignment='top', fontsize = 15)

# Plot max intensity
ax = fig.add_subplot(spec[1,0])
ax.scatter(lengths*(step_size/1000), Y_max, c='black')
ax.set_xlim(left=0)
ax.set_ylim(bottom=0)   
ax.set_xlabel('Microtubule length ($\mu$m)', fontsize=14)
ax.set_ylabel('Peak motor density (a.u.)' , fontsize=14)

# Plot FWHM of traffic jam
ax = fig.add_subplot(spec[1,1])
ax.scatter(lengths* (step_size/1000), np.multiply(Y_width,1000), c='black')
ax.set_xlabel('Microtubule length ($\mu$m)', fontsize=14)
ax.set_ylabel('Width traffic jam (nm)' , fontsize=14)

plt.show()
```

### Plot motor protein densities for multiple concentrations

```python
print('Available concentrations for SP dynein: ' + str(np.sort(RESULTS[RESULTS.Type.eq('SP')].Concentration.unique())))
print('Available concentrations for WT dynein: ' + str(np.sort(RESULTS[RESULTS.Type.eq('WT')].Concentration.unique())))
```

```python
# Select concentrations to plot
dynein_conc_range = [4, 1.5, 0.5, 0.2, 0.075, 0.025]  
# dynein_conc_range = [10, 2.5, 0.6, 0.15, 0.04, 0.01]
# dynein_conc_range = [8 2.7 1 0.3 0.1 0.03]

# Select dynein type: 'SP' or 'WT'
dynein_type = 'SP'

# Option to convolve the dynein profiles with an experimental PSF
convolve_data = True
```

```python
# Plot steady-state dynein profile for all lattice lengths

# Setup figure
cols = 2
rows = int(np.ceil(len(dynein_conc_range)/2))
fig = plt.figure(constrained_layout=True, figsize=(cols*6 ,rows*3))
spec = fig.add_gridspec(ncols=cols, nrows=rows)

fig_num = 0

for row in range(rows):
    for col in range(cols):         
        if fig_num < len(dynein_conc_range):            
            data_filter = RESULTS[RESULTS.Type.eq(dynein_type) & RESULTS.Concentration.eq(dynein_conc_range[fig_num])]
            lengths = data_filter.Length.unique()
            colors = [cm.copper(x) for x in np.arange(0, 256, len(lengths))]
            Y_max = []
            ax = fig.add_subplot(spec[row, col])    
            for length, color in zip(lengths, colors):
                
                # Prepare data
                data_set = RESULTS[RESULTS.Type.eq(dynein_type) & 
                                   RESULTS.Concentration.eq(dynein_conc_range[fig_num]) & 
                                   RESULTS.Length.eq(length)]
                final_state = []
                for data, time in zip(data_set.Lattice, data_set.Frametime):                 
                    if time == 0:
                        final_state.append(data)
                    elif time > 0:
                        final_state.append(data[-1])        

                if convolve_data:
                    Y = np.mean(final_state, axis=0)   
                    Y = np.concatenate((np.zeros(sz), Y, np.zeros(sz)), axis=None) 
                    Y = np.convolve(Y, gauss_filter, 'same')
                    X = np.arange(-sz, len(Y)-sz, 1) * (step_size/1000) 
                else:
                    Y = np.mean(final_state, axis=0)  
                    X = np.arange(0, len(Y), 1) * (step_size/1000)   
                
                Y_max.append(np.max(Y))
                ax.plot(X, np.flip(Y), color=color, linewidth=2)

            ax.set_xlim((-2, 10))        
            ax.set_xlabel('Microtubule length ($\mu$m)', fontsize=13)
            ax.set_ylabel('Motor density (a.u.)' , fontsize=13)
            ax.text(9, np.max(Y_max), str(dynein_type) + ': ' +str(dynein_conc_range[fig_num]) + ' nM', 
                    horizontalalignment='right', verticalalignment='top', fontsize = 14)
            
            fig_num += 1            
            
plt.show()
```

### Plot plus-end slope for multiple concentrations
Only available for the unconvolved dataset

```python
# Function to fit the plus-end slope for unconvolved dataset
def slopefit(x, y, start=10):
    rsquared = []
    coeffs = []
    result = []
        
    for i in range(len(y)-start):
        fit_window = np.arange(0, start+i, 1)
        X = x[fit_window]
        Y = y[fit_window]
        coeffs.append(np.polyfit(X, Y, 1))
        
        # Calculate R-squared
        p = np.poly1d(coeffs[-1])
        # fit values, and mean
        yhat = p(X)                      # or [p(z) for z in x]
        ybar = np.sum(Y)/len(Y)          # or sum(y)/len(y)
        ssreg = np.sum((yhat-ybar)**2)   # or sum([ (yihat - ybar)**2 for yihat in yhat])
        sstot = np.sum((Y - ybar)**2)    # or sum([ (yi - ybar)**2 for yi in y])
        rsquared.append(ssreg / sstot)

    index_max = np.argmax(rsquared)
    result = np.round(coeffs[index_max],3)
    
    return result
```

```python
# Short lattive sizes are difficult to fit, so length restriction is advised
length_cutoff = 2500 # in nm

fig, ax = plt.subplots(figsize=(7,7))
for conc in dynein_conc_range:
    data_filter = RESULTS[RESULTS.Type.eq(dynein_type) & RESULTS.Concentration.eq(conc)]
    lengths = data_filter.Length.unique()
    lengths = lengths[lengths > (length_cutoff/step_size)] 
    slope = []
    for length in lengths:
        # Prepare data        
        data_set = RESULTS[RESULTS.Type.eq(dynein_type) & 
                           RESULTS.Concentration.eq(conc) & 
                           RESULTS.Length.eq(length)]
        final_state = []
        for data, time in zip(data_set.Lattice, data_set.Frametime):                 
            if time == 0:
                final_state.append(data)
            elif time > 0:
                final_state.append(data[-1])        

        Y = np.mean(final_state, axis=0)  
        X = np.arange(0, len(Y), 1) * (step_size/1000)
        
        slope.append(slopefit(X, Y)[0])
    
    ax.errorbar(conc, np.median(slope), yerr=np.std(slope), fmt='o', c='black')
ax.set_xlabel('Concentration (nM)', fontsize=13)
ax.set_ylabel('Plus-end slope ($\mu$m$^{-1}$)' , fontsize=13)
plt.show()
```
